```python
import cv2
import numpy as np
import os
import glob
from PIL import Image
import matplotlib.pyplot as plt
```

```python
def crop_image(img):
    cropped_img = img[:1035,400:1435,:]
    return cropped_img
```

```python
def write_image_to_file(img, path):
    cv2.imwrite(path, img)
```

```python
def convert_video_to_frames(read_path, save_path):
    vidcap = cv2.VideoCapture(read_path)
    success, image = vidcap.read()
    count = 0
    while success:
        cropped_img = crop_image(image)
        cv2.imwrite("{}/frame_{:05d}.png".format(save_path, count), cropped_img)     # save frame as JPEG file
        success, image = vidcap.read()
        print('Read a new frame: ', count)
        count += 1
        if (count > 5350):
            break
```

```python
def read_image(path):
    img = np.asarray(Image.open(path))
    return img
```

```python
def read_image_directory(path):
    files = sorted([os.path.join(path, x) for x in os.listdir(path) if not x.startswith('.')])
    images = []
    for x in files:
        print(x)
        images.append(read_image(x))
    return images
```

```python
def resize_image(img, resize_shape):
    res = cv2.resize(img, dsize=resize_shape, interpolation=cv2.INTER_CUBIC)
    return res    
```

```python
images = read_image_directory(read_path)
```

```python
def train_bg_removal(start, end):
    backSub = cv2.createBackgroundSubtractorMOG2()
    masks = []
    for i in range(start, end):
        frame = images[i]
        fgMask = backSub.apply(frame)
        masks.append(fgMask)
    return backSub, masks
```

```python
def predict_bg_removal(start, end, backSub):
    masks = []
    for i in range(start, end):
        frame = images[i]
        fgMask = backSub.apply(frame)
        masks.append(fgMask)
    return masks
```
